```python
import pandas as pd

# Lista de estudiantes
estudiantes = []

while True:
    print("\n--- Menú Notas Universidad ---")
    print("1. Agregar estudiante")
    print("2. Mostrar estudiantes")
    print("3. Mostrar tabla de resultados (pandas)")
    print("4. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre del estudiante: ")
        nota = float(input("Nota final: "))
        estudiantes.append((nombre, nota))  # Guardamos como tupla
    elif opcion == "2":
        print("\nListado de Estudiantes:")
        for estudiante in estudiantes:
            print(f"{estudiante[0]} - Nota: {estudiante[1]}")
    elif opcion == "3":
        if estudiantes:
            # Convertir a DataFrame
            df = pd.DataFrame(estudiantes, columns=["Nombre", "Nota"])
            # Clasificar como Gana o Pierde
            df["Estado"] = df["Nota"].apply(lambda x: "Gana" if x >= 3.0 else "Pierde")
            print("\nTabla de Resultados:")
            print(df)

        else:
            print("No hay estudiantes registrados.")
    elif opcion == "4":
        print("¡Hasta luego!")
        break
    else:
        print("Opción inválida.")
```
