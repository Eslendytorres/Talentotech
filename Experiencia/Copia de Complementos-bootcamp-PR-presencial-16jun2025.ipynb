# Manejo de errores con excepciones (try...except):

````python
print("Ejemplo (división) de Manejo de errores con excepciones (try...except):")
num = float(input("Ingrese el numerador: "))
den = float(input("Ingrese el denominador: "))

try:
  div = num/den
  print(f"El resultado de la división es: {div}")
except ZeroDivisionError:
  print("Recuerda que no es posible dividir entre cero!")
````

# Función lambda:

Ver más ejemplos en: https://www.w3schools.com/python/python_lambda.asp

````python
n1 = float(input("Enter the 1st number: "))
n2 = float(input("Enter the 2nd number: "))

suma = lambda a,b: a+b
resultado = suma(n1, n2)
print(f"The sum is: {resultado}")

resta = lambda a,b: a-b
resultado = resta(n1, n2)
print(f"The substraction is: {resultado}")

multiplicacion = lambda a,b: a*b
resultado = multiplicacion(n1, n2)
print(f"The multiplication is: {resultado}")

division = lambda a,b: a/b
resultado = division(n1, n2) if n2!=0 else "You cannot divide by zero"
print(f"The division is: {resultado}")
````

````python
# TAREA (terminar siguiendo el ejemplo anterior):
import math

# 1. Operaciones de potencia y raíz
potencia = lambda base, exponente: math.pow(base, exponente)
raiz = lambda numero, indice=2: math.pow(numero, 1/indice)  # Por defecto raíz cuadrada

# 2. Operaciones logarítmicas
logaritmo = lambda numero, base=math.e: math.log(numero, base)
log10 = lambda numero: math.log10(numero)

# Ejemplos de uso:
print("5 elevado a 3:", potencia(5, 3))          # 125.0
print("Raíz cuadrada de 16:", raiz(16))         # 4.0
print("Raíz cúbica de 64:", raiz(64, 3))        # 4.0
print("Log natural de 10:", logaritmo(10))       # ~2.302
print("Log base 10 de 100:", log10(100))        # 2.0
````

````python
# TAREA (terminar siguiendo el ejemplo anterior):
import math
# 1. Funciones trigonométricas básicas
seno = lambda angulo_radianes: math.sin(angulo_radianes)
coseno = lambda angulo_radianes: math.cos(angulo_radianes)
tangente = lambda angulo_radianes: math.tan(angulo_radianes)
import math
# 2. Funciones trigonométricas inversas
arcsen = lambda valor: math.asin(valor)
arccos = lambda valor: math.acos(valor)
arctan = lambda valor: math.atan(valor)

# Ejemplos de uso:
angulo = math.pi/4  # 45 grados en radianes
print("Seno de π/4:", seno(angulo))          # ~0.707
print("Coseno de π/4:", coseno(angulo))      # ~0.707
print("Tangente de π/4:", tangente(angulo))  # ~1.0

print("Arcoseno de 0.5:", arcsen(0.5))       # ~0.523 rad (30°)
print("Arcocoseno de 0.5:", arccos(0.5))     # ~1.047 rad (60°)
print("Arcotangente de 1:", arctan(1))       # ~0.785 rad (45°)
````

```
>>> def fibonacci(n):
```

````python
def fibonacci(n): # n debe ser un numero entero positivo incluyendo el cero (numero natural)
     if n <= 1:
         return n
     else:
         return fibonacci(n - 1) + fibonacci(n - 2) # aquí se aplica la recursividad pues dentro del mismo bloque de código llamo a la función

print(fibonacci(0)) # retorna 0
print(fibonacci(1)) # retorna 1
print(fibonacci(2)) # retorna la suma de los dos anteriores: 1 + 0 = 1
print(fibonacci(3)) # retorna la suma de los dos anteriores: 1 + 1 = 2
print(fibonacci(4)) # complete ud!
print(fibonacci(5)) # complete ud!
print(fibonacci(6)) # complete ud!
print(fibonacci(7)) # complete ud!
print(fibonacci(8)) # complete ud!
print(fibonacci(9)) # complete ud!
print(fibonacci(10)) # complete ud!

# TAREA: hacer un ciclo while/for hasta 100: fib(n=100) = fib(n=99) + fib(n=98)

# debe llegar uno a uno hasta:

# Para n grandes es muy inficiente
#*print(fibonacci(98)) # complete ud!
#*print(fibonacci(99)) # complete ud!
#*print(fibonacci(100)) # fib(n=100) = fib(n=99) + fib(n=98)

# Como la recursividad puede consumir tiempo y memoria, hacerlo solo hasta n=20
# Consulta: cómo volver más eficiente las funciones recursivas?

print(fibonacci(11)) # complete ud!
print(fibonacci(12)) # complete ud!
print(fibonacci(13)) # complete ud!
print(fibonacci(14)) # complete ud!
print(fibonacci(15)) # complete ud!
print(fibonacci(16)) # complete ud!
print(fibonacci(17)) # complete ud!
print(fibonacci(18)) # complete ud!
print(fibonacci(19)) # complete ud!
print(fibonacci(20)) # complete ud!
print(fibonacci(21)) # complete ud!
print(fibonacci(22)) # complete ud!
print(fibonacci(23)) # complete ud!
print(fibonacci(24)) # complete ud!
print(fibonacci(25)) # complete ud!

# TAREA: HACER LA FUNCIÓN FACTORIAL RECURSIVA Y NO RECURSIVA (más eficiente)
````
