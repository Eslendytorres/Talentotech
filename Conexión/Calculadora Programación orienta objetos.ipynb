```python
import math

# Definimos la clase Calculadora
class Calculadora:
    def __init__(self, numero1, numero2):
        self.numero1 = numero1
        self.numero2 = numero2

    def sumar(self):
        return self.numero1 + self.numero2

    def restar(self):
        return self.numero1 - self.numero2

    def multiplicar(self):
        return self.numero1 * self.numero2

    def dividir(self):
        if self.numero2 != 0:
            return self.numero1 / self.numero2
        else:
            return "❌ Error: división por cero."

    def logaritmo(self):
        if self.numero1 > 0:
            return math.log10(self.numero1)
        else:
            return "Error: el logaritmo solo está definido para números positivos."

    def potencia(self):
        return math.pow(self.numero1, self.numero2)

    def raiz(self):
        if self.numero1 >= 0:
            return math.sqrt(self.numero1)
        else:
            return "Error: la raíz cuadrada de un número negativo no está definida en los reales."

# Menú interactivo
while True:
    print("\n MENÚ DE CALCULADORA")
    print("1) Suma")
    print("2) Resta")
    print("3) Multiplicación")
    print("4) División")
    print("5) Logaritmo (log10 del primer número)")
    print("6) Potencia (num1 elevado a num2)")
    print("7) Raíz cuadrada (de num1)")
    print("8) Salir")

    opcion = int(input(" Ingrese una opción (1-8): "))

    if opcion == 8:
        print(" ¡Gracias por usar la calculadora!")
        break

    # Pedir números solo si la opción es válida
    try:
        num1 = float(input(" Ingrese el primer número: "))
        if opcion != 7 and opcion != 5:  # Raíz y log solo necesitan un número
            num2 = float(input(" Ingrese el segundo número: "))
        else:
            num2 = 0  # Valor por defecto
    except ValueError:
        print("⚠️ Entrada inválida. Debes escribir números.")
        continue

    # Crear objeto de la clase Calculadora
    mi_calculadora = Calculadora(num1, num2)

    # Ejecutar según la opción
    if opcion == 1:
        print(" Resultado:", mi_calculadora.sumar())
    elif opcion == 2:
        print(" Resultado:", mi_calculadora.restar())
    elif opcion == 3:
        print(" Resultado:", mi_calculadora.multiplicar())
    elif opcion == 4:
        print(" Resultado:", mi_calculadora.dividir())
    elif opcion == 5:
        print(" Resultado:", mi_calculadora.logaritmo())
    elif opcion == 6:
        print(" Resultado:", mi_calculadora.potencia())
    elif opcion == 7:
        print(" Resultado:", mi_calculadora.raiz())
    else:
        print(" Opción inválida. Intenta de nuevo.")
```
