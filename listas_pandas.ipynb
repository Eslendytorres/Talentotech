```python
import pandas as pd

# Create a dataset
# Let's build the values lists
value_n_list = ['Andrew', 'Phillip', 'John', 'Laura', 'Sophia', 'Michael', 'Emma', 'Daniel', 'Olivia', 'James', 'Liam', 'Isabella', 'Lucas']
value_w_list = [50, 78, 91.5, 60, 55.2, 85.4, 48.3, 74.1, 62.0, 80.5, 77.3, 50.0, 70.2]  # [kg]
value_h_list = [1.76, 1.87, 1.69, 1.65, 1.70, 1.80, 1.58, 1.75, 1.68, 1.85, 1.79, 1.60, 1.77]  # [m]
# value_bmi_list will be calculated: bmi = w/(h^2)
# value_bmi_list = [25]*13 # init values [kg/m2]

# Let's use a dictionary ({key:value})
# with the format {"str":list}
data_dict = {"name": value_n_list,
             "weight": value_w_list,
             "height": value_h_list  #,
             # "BMI": value_bmi_list
             }
# Turn a dictionary into a dataframe
df = pd.DataFrame(data_dict)

# calculate bml
df['BMI'] = df['weight'] / (df['height'] ** 2)

def classify_bmi(bmi):
    if bmi < 18.5:
        return "Bajo peso"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

df['Estado de peso'] = df['BMI'].apply(classify_bmi)

df
```

```python
import pandas as pd

# Create a dataset
value_n_list = ['Andrew', 'Phillip', 'John', 'Laura', 'Sophia', 'Michael', 'Emma', 'Daniel', 'Olivia', 'James', 'Liam', 'Isabella', 'Lucas']
value_w_list = [50, 78, 91.5, 60, 55.2, 85.4, 48.3, 74.1, 62.0, 80.5, 77.3, 50.0, 70.2]  # [kg]
value_h_list = [1.76, 1.87, 1.69, 1.65, 1.70, 1.80, 1.58, 1.75, 1.68, 1.85, 1.79, 1.60, 1.77]  # [m]

# Crear el diccionario inicial
data_dict = {
    "name": value_n_list,
    "weight": value_w_list,
    "height": value_h_list
}

# Crear el DataFrame
df = pd.DataFrame(data_dict)

# Calcular el IMC
df['BMI'] = df['weight'] / (df['height'] ** 2)

# Clasificar el estado de peso
def classify_bmi(bmi):
    if bmi < 18.5:
        return "Bajo peso"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

df['Estado de peso'] = df['BMI'].apply(classify_bmi)

# Mostrar la tabla final
print(df)
df
```
